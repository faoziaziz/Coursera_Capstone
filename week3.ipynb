```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

*Another way, which would help to learn for more complicated cases of web scraping is using the BeautifulSoup package. Here is the package's main documentation page: http://beautiful-soup-4.readthedocs.io/en/latest/*


```python
import requests
from bs4 import BeautifulSoup
# GET Website url

website_url=requests.get(url).text
```

```python

soup = BeautifulSoup(website_url, 'html.parser')
My_table = soup.find('table', {'class':'wikitable'})
table_rows=My_table.find_all('tr')
# Just read from table 
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
```

```python
# initlaization data_frame
column_names = ["Postal Code", "Borough", "Neighborhood"]
df=pd.DataFrame(columns=column_names)

# add data to dataframe 
k=0
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    
    if k>=0:
        try:
            #print(len(row))
            df.loc[-1] = row  # adding a row
            df.index = df.index + 1  # shifting index
        except:
            print("gak bisa masuk")
            pass
    k+=1
# view dataframe
df
```

```python
df = df.replace('\n',' ', regex=True)
df.head()


```

```python
# get shape

df.shape
```

## rearrange and renaming column

```python
df.columns = ['Borough', 'Neighbourhood','Postcode']

cols = df.columns.tolist()
cols

cols = cols[-1:] + cols[:-1]

df = df[cols]

df.head()
```

#### Removing any space in the start of the string

```python
df['Borough'] = df['Borough'].str.strip()
```

#### Dropping all cells with a borough that is Not assigned

```python
df.drop(df.index[df['Borough'] == 'Not assigned'], inplace = True)

# Reset the index and dropping the previous index
df = df.reset_index(drop=True)

df.head(10)
```

#### Combining Neighbourhoods based on similar Postcode and Borough  

```python
df = df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(','.join).reset_index()
df.columns = ['Postcode','Borough','Neighbourhood']
df.head(10)
```

#### Assigning Borough values to the Neignbourhood where vlaue is "Not assigned"

i assign borough value and save it to df_web1.csv before i remove the bot assigment


```python
df.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df['Borough']
df.to_csv(r'df_web1.csv')
```
